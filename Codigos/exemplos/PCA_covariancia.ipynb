# Exemplo PCA 

```python
import pandas as pd
import numpy as np
#import 
```

```python
#Carregando o banco de dados
BD_decathon = pd.read_csv("../../Banco/decathlon(1).csv", index_col=0)
BD_decathon.head()
```

```python
pd.DataFrame(np.matrix(BD_decathon.iloc[:,0:10]))
```

Reduzindo o banco

```python
BD_decathon = BD_decathon.drop(['Rank','Points','Competition'], axis=1)
```

## Matriz de dados e Matriz de covariância

```python
#Matriz de dados
X = pd.DataFrame(np.matrix(BD_decathon))
S = X.cov()
print("Matriz de dados")
print(X)
print("\n")
```

```python
print("Matriz de covariancia")
print(S)
```

## Componentes principais

```python
#Autovalores e autovetores
autovalues, autovectors = np.linalg.eig(S)
autovectors
```

```python
autovector_pd = pd.DataFrame(autovectors)
autovector_pd
```

```python
#Scores
autovector_pd.dot(X.T)
```
