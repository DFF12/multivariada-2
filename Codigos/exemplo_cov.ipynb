## Matrix X e matriz de covariancia

```python
X = np.matrix(BD_dec.iloc[:,0:10]) #outro modo de extrair as 10 primeiras colunas do banco de dados
print("Matriz X:")
print(X)
S = np.cov(np.transpose(X))
print("Matriz de Covariância:")
print(S)
```

# Componentes principais

Obtem os scores

```python
pca_decathon=PCA(n_components=10)
pca_decathon.fit(np.asarray(X))
scores = pca_decathon.transform(np.asarray(X))
```
